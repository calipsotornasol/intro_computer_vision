# Solucion

Un arreglo de dos dimensiones se puede revertir así:

```python
import numpy as np
a = np.array([[10, 1, 1],
               [2, 0, 2],
               [3, 3, 3]])
              
```

```python
a
```

```python
a[:,::-1]
```

O así

```python
a[::-1]
```

Luego si queremos revertir una imagen podemos usar la misma idea

```python
import os
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
im = np.array(Image.open('data/empire.jpg'))
```

```python
plt.imshow(im);
```

```python
plt.imshow(im[:,::-1]);
```

```python
plt.imshow(im[::-1]);
```
